```python
import numpy as np
import pandas as pd
```

```python
#1. Leer la base de datos
dfP = pd.read_csv('C:/Users/camil/Desktop/k1000/Notebooks/Data Wrangling/p_data.csv', sep = ';')
dfW = pd.read_csv('C:/Users/camil/Desktop/k1000/Notebooks/Data Wrangling/w_data.csv', sep = ';')
dfP
```

```python
#2. Hacer merge de las dos bases de datos usando el id
df = pd.merge(dfP, dfW, on = 'ID')
df.set_index('ID')
df
```

```python
#3. Contar duplicados ////
df.groupby("ID").size().sort_values(ascending=False)
```
